```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("heart.csv")
df
```

## preprocessing ##

```python
df.dropna(inplace=True)

lable_encoder={}
for col in df.select_dtypes(include=['object']).columns:
    le=LableEncoder()
    df[col]=le.fit_transform(df[col])
    lable_encoder[col]=le
```

## spliting dataset into training and testing sets##

```python
x=df.drop(columns=['target'])
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Standardize features ##

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

## Trainn and Evaluate SVM Models ## 

```python
svm_linear=SVC(kernel='linear',C=1)
svm_linear.fit(x_train,y_train)
y_pred_linear=svm_linear.predict(x_test)


svm_ploy=SVC(kernel='poly',C=1,degree=3)
svm_ploy.fit(x_train,y_train)
y_pred_poly=svm_ploy.predict(x_test)


svm_rbf=SVC(kernel='rbf',C=1,gamma='scale')
svm_rbf.fit(x_train,y_train)
y_pred_rbf=svm_rbf.predict(x_test)
```

## Evaluate Model Performance ##

```python
print(f'Linear Kernel Accuracy: {accuracy_score(y_test,y_pred_linear):.2f}')
print(f'polynomial Kernel Accuracy: {accuracy_score(y_test,y_pred_poly):.2f}')
print(f'RBF  Kernel Accuracy: {accuracy_score(y_test,y_pred_rbf):.2f}')
```
